# Read Excel File and Perform Various Activities on File

```python
#Read File

import xlrd
location = ("E:\Development\PIAIC\OpenDataPakistan\public-census_oct_2018.xlsx")
workBook = xlrd.open_workbook(location)
sheet = workBook.sheet_by_index(0)
sheet.cell_value(0,0)

#File Name

print("Sheet Name is: ", workBook.sheet_names(),end='')
print('\n')

#number of rows

print("Total no of Rows: ", sheet.nrows, end='')
print('\n')

#number of columns

print("Total no of columns: ",sheet.ncols, end='')
print('\n')

#columns name

for i in range(sheet.ncols):
    print("Column # ", i,"is: ", sheet.cell_value(0,i), end=',')
    print('\n')

#print column values
colValue = int(input("Please enter column number: "))
for i in range(sheet.nrows):
    print(sheet.cell_value(i,colValue), end='')
    print('\n')

#print row values
rowValue = int(input("Please enter row number: "))
for i in range(sheet.ncols):
    print(sheet.cell_value(rowValue,i), end=',')
    
listData = workBook.sheet_name().range('A2:A1231').value
listData
```
